O objetivo deste projeto é criar um dashboard com os dados deste [repositório](https://github.com/TheMrityunjayPathak/FinanceReportDashboard).

Os dados estão sujos, então será necessário realizar o tratamento (abaixo) e para os valores futuros pode automatizá-los facilmente em um script para automáticamente executar as alterações.

```python
link = 'https://raw.githubusercontent.com/TheMrityunjayPathak/FinanceReportDashboard/main/financial_data.csv'
import pandas as pd
df = pd.read_csv(link)
```

```python
df.head(3)
```

```python
df['Country'] = df['Country'].str.replace('United States of America', 'USA')
df.columns = df.columns.str.strip() # Remove espaços desnecessarios nas colunas
```

```python
# Alterando valores nas variaveis
df['Units Sold'] = df['Units Sold'].str.replace('$', '') # Retirando caracteres das variaveis numéricas
df['Units Sold'] = df['Units Sold'].str.strip() # Remove espaços desnecessarios nas colunas
df['Units Sold'] = df['Units Sold'].str.replace(',', '').astype(float) # Substituir vírgulas por pontos e converter para float
df['Units Sold'] = df['Units Sold'].astype("float64")
```

```python
# Fazendo o mesmo processo em colunas semelhantes
import numpy as np
colunas = ['Manufacturing Price',	'Sale Price', 'Gross Sales', 'Sales',	'COGS']
# Iterar sobre as colunas
for i in colunas:
    # Substituir vírgulas por pontos e converter para float
    df[i] = df[i].astype(str)
    df[i] = df[i].replace('-', '')
    df[i] = df[i].str.replace('$', '')
    df[i] = df[i].str.replace(',', '')
    df[i] = df[i].replace('-', np.nan)
    df[i] = df[i].str.strip().astype(float)
```

A coluna Discounts tem alguns problemas extras, vamos trata-los:

```python
df['Discounts'] = df['Discounts'].str.strip().replace('$-', '')
df['Discounts'] = df['Discounts'].str.strip()
```

```python
df['Discounts'] = df['Discounts'].str.replace('$', '')
df['Discounts'] = df['Discounts'].str.strip()
```

```python
df['Discounts'] = df['Discounts'].str.replace(',', '')
df['Discounts'] = df['Discounts'].str.strip()
```

```python
# Removendo linhas com espaços em branco para valores com 0
df['Discounts'].replace(r'^\s*$', np.nan, regex=True, inplace=True)
df.fillna(0, inplace=True)
```

```python
df['Discounts'] = df['Discounts'].astype(float)
```

A coluna 'Profit' tem um problema adicional, os valores negativos além de conterem os caracteres extras contem os parenteses. É necessário removê-los sem remover os numeros de dentro e transformá-los em negativo e depois em float64.

```python
# Mesmos problemas da coluna Discounts indentificados na coluna 'Profit'
df['Profit'] = df['Profit'].str.strip().replace('$-', '')
df['Profit'] = df['Profit'].str.strip()
df['Profit'] = df['Profit'].str.replace('$', '')
df['Profit'] = df['Profit'].str.strip()
df['Profit'] = df['Profit'].str.replace(',', '')
df['Profit'] = df['Profit'].str.strip()
# Removendo linhas com espaços em branco para valores com 0
df['Profit'].replace(r'^\s*$', np.nan, regex=True, inplace=True)
df.fillna(0, inplace=True)
```

```python
# Remover os parênteses externos coloca o sinal negativo e converte para float
df['Profit'] = df['Profit'].str.replace(r'\((.*?)\)', r'-\1', regex=True).astype(float)
df['Profit'] = df['Profit'].astype(float)
```

```python
df['Profit'] = df['Profit'].fillna(0)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```
